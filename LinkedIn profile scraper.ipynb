```python
#getting profile links
links = []
for page in range(6):
    sleep(5)
    profile = driver.find_elements_by_xpath("//a[@class='app-aware-link']")
    for i in range(len(profile)):
        profile = driver.find_elements_by_xpath("//a[@class='app-aware-link']")
        links.append(profile[i].get_attribute('href'))
    
    wait_until(Text('LinkedIn Corporation © 2022').exists)
    click('Next')
```

```python
# Data scraping
n = []
j = []
loc = []
left = []
right = []
link_1 = []
link_2 = []
link_3 = []
for l in range(10):
    go_to(links[l])
    src = driver.page_source
    soup = BeautifulSoup(src,'lxml')
    
    try:
        #profile datas
        name = soup.find('h1',{'class':'text-heading-xlarge inline t-24 v-align-middle break-words'}).text
        job = soup.find('div',{'class':'text-body-medium break-words'}).text.strip()
        location = soup.find('span',{'class':'text-body-small inline t-black--light break-words'}).text.strip()
        l_info = soup.find('ul',{'class':'pv-text-details__right-panel'}).text.strip().replace('\n','')
        r_info = soup.find('ul',{'class':'pv-top-card--list pv-top-card--list-bullet display-flex pb1'}).text.strip().replace('\n','')
        n.append(name)
        j.append(job)
        loc.append(location)
        left.append(l_info)
        right.append(r_info)
    except:
        n.append('None')
        j.append('None')
        loc.append('None')
        left.append('None')
        right.append('None')
    try:    
        click('Contact info')
        wait_until(Text('Contact Info').exists)
        sleep(10)
        contact_url = driver.find_element_by_xpath("//div[@class='pv-profile-section__section-info section-info']")
        url = contact_url.find_elements_by_tag_name('a')
        try:
            link_1.append(url[0].text)
        except:
            link_1.append('')
        try:
            link_2.append(url[1].text)
        except:
            link_2.append('')
        try:
            link_3.append(url[2].text)
        except:
            link_3.append('')
    except:
        link_1.append('')
        link_2.append('')
        link_3.append('')
            
        
```

```python
import pandas as pd
data = pd.DataFrame({'Name':n,'Job':j,'Location':loc,'Education':left,'Followers':right,'Link 1':link_1,'Link 2':link_2,
'Link 3':link_3})
```
